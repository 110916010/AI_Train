```python
import cv2
import matplotlib.pyplot as plt

def histogram(i):
    h = cv2.calcHist([i],[0],None,[256],[0,256])
    plt.plot(h)
    plt.xlim([0,256])
    plt.xlabel('Intensity')
    plt.ylabel('Number')
    plt.show()


def main():
    img = cv2.imread('myBaby.jpg')
    cv2.imshow('source',img)
    histogram(img)

main()
```
